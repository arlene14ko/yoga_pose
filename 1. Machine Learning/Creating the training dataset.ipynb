# Creating the dataset

```python
import cv2
import mediapipe as mp
import numpy as np
import os
import pandas as pd
import csv
mp_pose = mp.solutions.pose
```

```python
d=[]
for i in os.listdir('datasets/training data/training frames'):
    for f in os.listdir('datasets/training data/training frames/'+i):
        d.append((f'{i}', f'datasets/training data/training frames/{i}/{f}'))
```

## Dataset with Visibility

```python
landmarks = ['class']
for val in range(33):
    landmarks += ['x{}'.format(val), 'y{}'.format(val), 'z{}'.format(val), 'v{}'.format(val)]

with open('dataset_with_visibility.csv', mode='w', newline='') as f:
    csv_writer = csv.writer(f, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
    csv_writer.writerow(landmarks)
    
for class_name, file in d:
    frame = cv2.imread(file)
    # Initiate holistic model
    with mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5) as pose:
        # Recolor image to RGB
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        
        # Make detection
        results = pose.process(image)
        # Extract Pose landmarks
        pose = results.pose_landmarks.landmark
        pose_row = list(np.array([[landmark.x, landmark.y, landmark.z, landmark.visibility] for landmark in pose]).flatten())
            
        # Append class name 
        pose_row.insert(0, class_name)

        # Export to CSV
        with open('dataset_with_visibility.csv', mode='a', newline='') as f:
            csv_writer = csv.writer(f, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
            csv_writer.writerow(pose_row) 
                        
```
